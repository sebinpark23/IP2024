```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

##### Scaling

```python
img = cv2.imread('pokemon.png')

res = cv2.resize(img, None, fx=5, fy=5, interpolation = cv2.INTER_CUBIC)

#height, width = img.shape[:2]
#res = cv2.resize(img,(5*width, 5*height), interpolation = cv2.INTER_CUBIC)
```

```python
cv2.namedWindow('image')
while(1):
    cv2.imshow('image',res)
    k = cv2.waitKey(1)&0xFF
    if k == 27:
        break
cv2.destroyAllWindows()  
```

##### Translation

```python
img = cv2.imread('pokemon.png',0)
rows,cols = img.shape[:2]
M = np.float32([[1,0,100],  # x축으로 100, y축으로 50
                [0,1,50]])
dst = cv2.warpAffine(img,M,(cols,rows))
```

```python
cv2.imshow('img',dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

##### Rotation

```python
img = cv2.imread('pokemon.png')
rows,cols = img.shape[:2]
M = cv2.getRotationMatrix2D((cols/2,rows/2),45,1) # 45도 돌려라(이미지 중심에서) -> 죄표 값을 바꿔주면 중심이 바뀜
dst = cv2.warpAffine(img,M,(cols,rows))
```

```python
img = cv2.imread('pokemon.png')
rows,cols = img.shape[:2]
M = cv2.getRotationMatrix2D((cols/4,rows/4),45,1) # 중심이 바껴서 돌아감
dst = cv2.warpAffine(img,M,(cols,rows))
```

```python
cv2.imshow('img',dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

##### Affine Transformation

```python
img = cv2.imread('pokemon.png')
rows,cols,ch = img.shape

pts1 = np.float32([[50,50],[200,50],[50,200]])

pts2 = np.float32([[10,100],[200,50],[100,250]])

M = cv2.getAffineTransform(pts1,pts2)

dst = cv2.warpAffine(img,M,(cols,rows))

plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```

##### Perspective Transformation

```python
# homework3 코드
img = cv2.imread('dsu2.jpg')
rows,cols,ch = img.shape
 
pts1 = np.float32([[178,239],[821,219],[818,511],[178,502]])  # 왼쪽 위부터 시계방향
pts2 = np.float32([[0,0],[600,0],[600,400],[0,400]])
 
M = cv2.getPerspectiveTransform(pts1,pts2)

dst = cv2.warpPerspective(img,M,(600,400))  # 전체 이미지

plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```
