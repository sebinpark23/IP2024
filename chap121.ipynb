```python
import numpy as np  #numpy라는 오픈 cv 가져옴
import cv2
# Load an color image in grayscale
#,0 추가하면 흑백
# 이미지 읽어옴
img = cv2.imread('birthday0205.jpg')  
```

```python
cv2.imshow('image',img) # 창을 생성하고 이미지 표시
cv2.waitKey(0)
cv2.destroyAllWindows() # 모든 OpenCV 창 닫음
```

```python
cv2.namedWindow('image', cv2.WINDOW_NORMAL) # 창 크기를 조절할 수 있도록 설정
cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.imwrite('messigray.png',img) # 이미지를 'messigray.png' 로 저장
```

```python
img = cv2.imread('birthday0205.jpg',0)
cv2.imshow('image',img)
k = cv2.waitKey(0)
if k == 27: # wait for ESC key to exit
    cv2.destroyAllWindows()
elif k == ord('s'): # s 키를 누르면
    cv2.imwrite('messigray.png',img) # 이미지 저장
    cv2.destroyAllWindows()
```
