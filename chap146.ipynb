##### Chap 1.4.6 Image Gradients

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('sIMG_2168.jpg',0)
 
laplacian = cv2.Laplacian(img,cv2.CV_64F)
sobelx = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=5)  # 왼쪽과 오른쪽 차이
sobely = cv2.Sobel(img,cv2.CV_64F,0,1,ksize=5)  # 밑에꺼와 아래 위로 차이
```

```python
plt.subplot(2,2,1),plt.imshow(img,cmap = 'gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])

plt.subplot(2,2,2),plt.imshow(laplacian,cmap = 'gray')
plt.title('Laplacian'), plt.xticks([]), plt.yticks([])  # 잘보면 격자보임(두번 미분) 자기 왼쪽 오른쪽 변화 

plt.subplot(2,2,3),plt.imshow(sobelx,cmap = 'gray')
plt.title('Sobel X'), plt.xticks([]), plt.yticks([])    # 왼 오 차이 때문에 세로 줄 있음

plt.subplot(2,2,4),plt.imshow(sobely,cmap = 'gray')
plt.title('Sobel Y'), plt.xticks([]), plt.yticks([])    # 위 아래라서 

# edge는 소벨 X, Y 합하기

plt.show()
```
